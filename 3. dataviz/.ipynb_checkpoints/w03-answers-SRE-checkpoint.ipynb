## 1. Downsample the data by reservoir sampling

Here I'm using straight python for reading the file. Modules like pandas give us functions to parse a data table from a file, but I like knowing I can fall back to basics. 

The business with reading a line, stripping the trailing newline, skipping blank lines and comments is practically idiomatic in my own work, because I often encounter (and use) text data files that have comments starting with '#' and blank lines, for human readability.

```python
m = 10                            # <m> is how many lines to sample
n = 0                             # <n> counts what line we're on, starting from 1
S = []                            # S[] is our sample of <m> lines; 0..m-1
for line in open('w03-data.tbl'):
    line = line.rstrip('\n')      # Remove trailing newline
    if not line:       continue   # Skip blank lines
    if line[0] == '#': continue   # Skip comments

    # Reservoir sampling algorithm
    n += 1 
    if n <= m:                       
        S.append(line)
    else:
        roll = np.random.randint(0,n)   # roll is 0..n-1, because np.random.randint() samples on half-open interval [a,b)
        if roll < m:
            S[roll] = line

for line in S:
    print(line)

```

So far, so good. Random sample looks like gene name in first column, followed by TPM data in all other columns.

Now I'll read the data into a Pandas dataframe too, though it's not tidy data yet. I use arguments to tell Pandas it's a whitespace-delimited file, with comment lines starting with `#` that need to be ignored for now, with no parseable header, and the first column consists of row (gene) names. I figured this out by reading the documentation for `read_table`, making notes in my python cheat-sheet notebook about which arguments I think I'll want to remember and use, plus trial and error. The Pandas `sample()` method gives us a random sample.

## 2. Outliers and validation

We were asked to find and print the lines with the smallest and largest values. This is an example of looking for outliers in a data file, as a way of detecting stuff that's gone awry. Here, we'll find that there's some lines that have "-999.0" for the expression values, which seems wrong (we're not expecting negative expression values).

We were also asked to find "bad" lines, and "bad" was left up to your judgement. Here we test each expression value for whether it's a real number (floating point number). (Incidentally, that test serves as an introductory example to Python's "exception handling" using `try` and `except`.) We find that some lines have "n.d." for expression values, which presumably means "not done" (missing data). Missing data placeholders are common in biological data.

```python
datafile    = 'w03-data.tbl'
n           = 0             # Counter for line numbers
initialized = False         # Gets set to True once we've seen 1st data line


# Function: is_float(s)
#    Tests whether string s can be converted to a float or not;
#    returns True or False.
#      ... an example of Python's exception handling with try/except.
#
def is_float(s):
    try:
        float(s)
        return True
    except:
        return False


for line in open(datafile):          # For each line in the input file...
    n += 1                           # Keep track of line numbers 1..
    line   = line.rstrip('\n')       # Remove the trailing newline
    if not line:       continue      # Skip blank lines
    if line[0] == '#': continue      # Skip comment lines

    fields = line.split()            # Split into fields on whitespace
    badline = False                  

    # Check that all fields after the gene name are floats
    for f in fields[1:]:
        if not(is_float(f)):
            print('UNEXPECTED "{}" on line {}:\n   {}'.format(f, n, line))
            badline = True
            break
    if badline: continue

    # Keep the lines with min, max values on them;
    # initialize on the first non-comment data line.
    linemax =  max([float(f) for f in fields[1:]])   # Danger! the fields[] are STRINGS;
    linemin =  min([float(f) for f in fields[1:]])   #  but we want min, max as NUMBERS.
                                                     #  so we do a list comprehension to convert.
    if not(initialized):
        nfields = len(fields)
        maxval  = linemax
        maxline = line                                  
        maxidx  = n                                     
        minval  = linemin
        minline = line
        minidx  = n
        initialized = True
    else:
        if len(fields) != nfields:
            sys.exit('bad line: different number of fields ({},{})\n{}'.format(nfields, len(fields), line))
        if linemax > maxval:
            maxval  = linemax
            maxline = line
            maxidx  = n
        if linemin < minval:
            minval  = linemin
            minline = line
            minidx  = n

print('')
print('max "{}" is on line {}:\n    {}'.format(maxval, maxidx, maxline))                
print('min "{}" is on line {}:\n    {}'.format(minval, minidx, minline))
```

The max of 126.7 seems ok as an actual TPM value, but -999.0 is bad. Now we've discovered that there's some lines that have 'n.d.', and some that have '-999.0' values.

## 3. Clean the data

Remove the lines that contain -999.0 and n.d. bad values, and create a new file `my-data-clean.tbl`.

Incidentally, this gives you an example of writing to a file, instead of just to standard output. If we're writing to a file, we have to be especially careful to close it when we're done, or you can get a confusing (and easily overlooked) bug: a `print` statement _buffers_ its output, rather than writing it immediately, and if the program ends while there's still data in the output buffer, those data are lost, and your file will be semi-mysteriously truncated. Calling `outf.close()` closes the file. Another way to do this is with Python's `with..as` construct, which we'll use in the next script.

```python
infile    = 'w03-data.tbl'
outfile   = 'my-data-clean.tbl'
outf      = open(outfile, 'w')

for line in open(datafile):          # For each line in the input file...
    line   = line.rstrip('\n')       # Remove the trailing newline
    if not line: continue            # Skip blank lines
    if line[0] == '#':               # Regurgitate comment lines as is
        print(line, file=outf)
        continue
    fields = line.split()            # Split into fields on whitespace for examination

    if (fields.count('n.d.')   > 0): continue
    if (fields.count('-999.0') > 0): continue

    print(line, file=outf)
    
outf.close()
print('Saved clean data file to {}'.format(outfile))
```

## 4. Tidy the data

In tidy data, each line corresponds to a single observation, and column headers are random variables (features of the data observation that take on some value), not values. Here, the input data table has _two_ column headers that correspond to things that should be fixed variables: whether the expression value is from wild type or mutant (let's call this variable 'genotype', taking values 'wt' or 'mut'), and whether the expression value is from a male (M) or female (F) sand mouse (let's call this fixed variable 'sex'). 

An 'observation', in RNA-seq expression data, is all the data from one sample: every gene becomes a variable (a column) too. So much for tidy data! We have a file with 20 observations (20 RNA-seq experiments, for 10 mutant and 10 wt sand mice, some of which are male and some female), and for each observation we have ~20,000 columns. Not terribly human readable now (and some editors will start to choke on lines that long, but it's 'tidy' from the perspective of data handling.

To get that, we have to transpose the data matrix: our columns will become rows (observations), and our rows (genes) will become columns. The easiest way to transpose is to read the entire dataset into memory, then print it back out in the right orientation.

Same deal here with closing the output file, by the way, but we'll use the `with...as` construct now. The `with...as` gives us a clean idiom that closes the output file automatically when we're done.

My strategy is to parse the data myself, and output a file that Pandas can read easily. Getting Pandas to deal with the original file seemed more fiddly.

```python
infile  = 'my-data-clean.tbl'
outfile = 'my-data-tidy.tbl'

# First, ingest all the data.
#
N_wt  = 0    # Number of wt sand mouse samples
N_mut = 0    #   ... number of mutant
N_col = 0    # total number of observations, N_wt + N_mut
data  = {}   # We'll read the data into a dict of arrays: data[genename] = tpm[0..Ncol-1]
with open(infile) as f:
    for line in f:
        fields   = line.split()

        # Parse comments, looking for the column header lines.
        if line[0] == '#':
            # genotype labels header
            if len(fields) > 1 and (fields[1] == 'wt' or fields[1] == 'mut'):
                genotype_labels = fields[1:]                  # genotype_labels is now a list of the column headers 'wt' or 'mut' for each column
                N_wt           = genotype_labels.count('wt')
                N_mut          = genotype_labels.count('mut')
            # sex labels header
            elif len(fields) > 1 and fields[1] == 'gene_name':
                sex_labels  = fields[2:]                      # sex_labels is now a list, of the same length, of 'M' vs. 'F' for each column
                N_col       = len(sex_labels)
                assert(N_wt + N_mut == N_col), "Headers don't match"   # Assumes condition header is already parsed
            continue
                

        # If we're not parsing a column header or comment - then this
        # is a data line, starting with a gene name, and containing
        # N_col tpm values. Store the whole thing in a dict of tpm arrays,
        # indexed by gene name. We haven't converted tpm to numbers - 
        # we're just keeping them as their original strings.
        genename = fields[0]
        tpm      = fields[1:]
        assert (len(fields) == N_wt + N_mut + 1), "Bad data line"
        if tpm.count('-999.0') > 0 : continue      # The original data set has bad lines with -999.0 values
        if tpm.count('n.d.')   > 0 : continue      #   ... and n.d. values. Skip these lines.

        data[genename] = tpm     # This associates the tpm data array 0..Ncol-1 with <genename>.

        
# Now, output in tidy format.
# Starting with our column headers: "genotype", "sex", and then a whole lot of gene names.
with open(outfile, 'w') as outf:  
    genes = sorted(data)      # <genes> is now a sorted list of the keys (gene names) in <data>
    print("{0:8s} {1:>3s} ".format("genotype", "sex"), end='', file=outf)
    for genename in genes:
        print("{0:>20s} ".format(genename), end='', file=outf)
    print('', file=outf)

    for i in range(N_col):    # Columns of the original data table are our observations (rows)
        print("{0:8s} {1:>3s} ".format(genotype_labels[i], sex_labels[i]), end='', file=outf)
        for genename in genes:
            print("{0:>20s} ".format( data[genename][i]), end='', file=outf)
        print('', file=outf)


print('Saved tidy data to {}'.format(outfile))
```

## 5. Visualize the raw expression data

The key idea we're going for here is to _always look at the raw data, not just a summary of it_. If the data are too large to look at all at once, downsample it. 

The mistake that Lestrade _et al._ made was to trust box plots. When you summarize data with a box plot, by mean and standard deviation, you're effectively assuming that your data are normally distributed -- or at least, that your data's distribution is unimodal and symmetric. These data are not unimodal, and that turns out to be the key to understanding what's going on in _Coriander_ mutants.

The 'answer', using Seaborn, is just a few lines of code, but it may have taken you a bunch of fiddling around to arrive at it (it did for me). Although the most tidy data format has one column per gene name (~20,000 columns), it's a lot easier to get Seaborn to make pretty plots for us (for this exercise) if we have just a single column "gene" that takes the 20,000 gene names as values. Pandas has a `melt` function that makes it easy to reformat the original tidy data into the other version of tidy data. Then we can downsample, selecting only rows corresponding to a randomly sampled list of 10 genes.

Given those data, Seaborn makes it relatively easy to plot any two variables (columns) in a tidy data file against each other -- here, 'genotype' (wt or mut) vs 'tpm' (expression level), specified by the `x="genotype"` and `y="tpm"` arguments. 

Seaborn's `catplot` makes it easy to do multiple plots like that in a grid, specified by the `col="gene"` argument: i.e. one graph for every value we see in the "gene" column.

Here I'm plotting jittered strip plots (`kind="strip", jitter=True`); to make swarm plots use `kind=swarm`.

The `hue` argument lets you color data points according to yet another column. `hue="sex"` means assign a different color to male vs. female data points.

```python
tidyfile     = 'my-data-tidy.tbl'

data         = pd.read_table(tidyfile, delim_whitespace=True)  
genes        = data.columns.values.tolist()[2:]  # A trick to get gene names as a list!
gene_sample  = np.random.choice(genes, 10)       # Random sample of 10 gene names

data = pd.melt(data, id_vars=['genotype', 'sex'], var_name="gene", value_name="tpm")
data = data.loc[data['gene'].isin(gene_sample)]

g = sns.catplot(data=data, x="genotype", y="tpm", hue="sex", col="gene", kind="strip", height=4, aspect=0.5, col_wrap=5, jitter=True)
g.set_axis_labels("", "TPM (transcripts per million)")
g.set_titles("{col_name}")
```

## Conclusions

The `hue` argument is the key in the visualization. We see that in wild type, sand mouse genes are differentially expressed in males versus females, so the expression levels are bimodally distributed. In the _Coriander_ mutant, this sex specific expression is lost, and expression levels are unimodally distributed around the average for males/females.
